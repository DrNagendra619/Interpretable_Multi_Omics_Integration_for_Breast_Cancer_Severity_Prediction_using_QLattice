## Interpretation of ROC Curve Plot (Cell 13)

The plot generated by `models[1].plot_roc_curve(test)` visualizes the Receiver Operating Characteristic (ROC) curve for the model on the *unseen test data*. The ROC curve is a fundamental tool for evaluating the performance of a binary classification model at various classification thresholds. It plots the True Positive Rate (TPR), also known as recall or sensitivity, against the False Positive Rate (FPR) at different threshold settings.

### Plot Components and Interpretation:
*   **X-axis (False Positive Rate - FPR):** This represents the proportion of negative instances that were incorrectly classified as positive. It's calculated as `FP / (FP + TN)`. A lower FPR is generally desirable.
*   **Y-axis (True Positive Rate - TPR):** This represents the proportion of positive instances that were correctly classified as positive. It's calculated as `TP / (TP + FN)`. A higher TPR is desirable.
*   **The Curve:** The curve itself shows the trade-off between TPR and FPR across all possible classification thresholds. Each point on the curve corresponds to a specific threshold. Moving along the curve, a classifier might prioritize correctly identifying more positive cases (higher TPR), but this often comes at the cost of incorrectly classifying more negative cases (higher FPR).
*   **Diagonal Line (Dashed Line):** This represents a random classifier (or a classifier that makes predictions purely by chance). A model whose ROC curve falls below this line is performing worse than random guessing. A curve that hugs the top-left corner indicates a highly accurate model.
*   **Area Under the Curve (AUC):** The AUC score (which we saw in Cell 10 for the test set was 0.656) is a scalar metric that summarizes the overall performance of the classifier across all possible thresholds. It represents the probability that the model will rank a randomly chosen positive instance higher than a randomly chosen negative instance. An AUC of 1.0 means perfect classification, while an AUC of 0.5 means random guessing.

### Insights from the ROC Curve Plot:
*   **Discriminative Power:** The shape of the curve indicates the model's ability to distinguish between the positive and negative classes. An ideal curve would go straight up the Y-axis to 1.0 (100% TPR) and then straight across the X-axis to 0.0 (0% FPR). Our model's curve will likely be above the random guessing line, reflecting the AUC of 0.656, but not very close to the top-left corner.
*   **Trade-off between TPR and FPR:** The curve allows us to visualize the impact of choosing different classification thresholds. If we want to achieve a very high TPR (capture most of the positive cases), we might have to accept a higher FPR (misclassify more negative cases). Conversely, if we want to keep FPR very low, our TPR might suffer.
*   **Confirmation of AUC:** The visual representation of the curve reinforces the AUC value. A curve that bows out significantly towards the top-left indicates good discrimination, while a curve closer to the diagonal line indicates less effective discrimination.
*   **Context of Low Recall:** Given the low recall (0.156) on the test set, the ROC curve plot helps understand this limitation. A low recall suggests that even at thresholds aimed at maximizing TPR, the model struggles to identify a large proportion of actual positive cases. The ROC curve will visually confirm that to achieve higher TPR, we would likely need to accept a very high FPR, making it difficult to find a good balance.

In summary, the ROC curve provides a comprehensive view of the classifier's performance across different operating points, highlighting its discriminative ability and the trade-offs involved in setting classification thresholds.

```python
def feature_correlations_map(train, target, correlation='pearson', scale_features=True, abs_values = False):

    from sklearn.decomposition import PCA
    import plotly.express as px

    target_abs = f"{target}_abs"

    data_corr = train.corr(correlation).drop(target, axis = 0)

    target_corr = data_corr[target].values.copy()
    data_corr = data_corr.drop(target, axis = 1)

    features = data_corr.columns

    if scale_features:
        from sklearn.preprocessing import StandardScaler
        x = data_corr.loc[:, features].values
        data_corr = StandardScaler().fit_transform(x)


    pca = PCA(n_components=2)
    principalComponents = pca.fit_transform(data_corr)
    principalDf = pd.DataFrame(data = principalComponents, columns = ['PC1', 'PC2'])

    principalDf[target] = target_corr
    principalDf[target_abs] = np.abs(target_corr)
    principalDf['Features'] = features

    color_feature = target

    if abs_values:
        color_feature = target

    fig = px.scatter(principalDf, x="PC1",
                    y="PC2", color=color_feature, hover_name="Features", width=1000, height=800)

    fig.add_annotation(text=f'Variance explained <br>PC1: {"{:.0%}".format(pca.explained_variance_ratio_[0])} <br>PC2: {"{:.0%}".format(pca.explained_variance_ratio_[1])}',
                    align='left',
                    showarrow=False,
                    xref='paper',
                    yref='paper',
                    x=0.98,
                    y=0.98,
                    bordercolor='black',
                    borderwidth=1)

    fig.show()
```
